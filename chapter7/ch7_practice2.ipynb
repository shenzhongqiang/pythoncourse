```python
df_company = df[df.pe>0].sort_values(by="pe").head(10)
```

```python
from pyecharts.charts import Bar, Pie, Grid
from pyecharts import options as opts
bar = Bar()
bar.add_xaxis(df_company.index.values.tolist())
bar.add_yaxis("PE", df_company.pe.values.tolist())
bar.add_yaxis("PB", df_company.pb.values.tolist())
bar.set_global_opts(title_opts=opts.TitleOpts(title="上市公司市盈率和市净率"))
bar.render_notebook()
```

```python
df_industry = df.groupby("industry")["industry"].agg({"company_num": "count"})
```

```python
from pyecharts import options as opts
from pyecharts.charts import Page, WordCloud
from pyecharts.globals import SymbolType

words = list(zip(df_industry.index.values.tolist(), df_industry.company_num.values.tolist()))
wc = WordCloud()
wc.add("", words, word_size_range=[20, 100])
wc.set_global_opts(title_opts=opts.TitleOpts(title="WordCloud-基本示例"))
wc.render_notebook()
```

```python
years = [2018, 2019]
quarters = [1, 2, 3, 4]
symbol = "000001"
net_profits = []
business_incomes = []
dates = []
for year in years:
    for quarter in quarters:
        df = ts.get_profit_data(year, quarter)
        if len(df[df.code=="000001"]) == 0:
            continue
        net_profit = df[df.code=="000001"].iloc[0].net_profits
        business_income = df[df.code=="000001"].iloc[0].business_income
        net_profits.append(net_profit)
        business_incomes.append(business_income)
        date = "{}-{}".format(year, quarter * 3)
        dates.append(date)
print(net_profits)
print(business_incomes)
print(dates)
```

```python
import pyecharts.options as opts
from pyecharts.faker import  Faker
from pyecharts.charts import Line

line = Line()
line.add_xaxis(dates)
line.add_yaxis("净利润", net_profits)
line.add_yaxis("营收", business_incomes)
line.set_global_opts(title_opts=opts.TitleOpts(title="净利润和营收"))
line.render_notebook()
```
