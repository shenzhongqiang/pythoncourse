```python
x = np.arange(100)
y = np.sin(x/50*np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks([0, 50, 100])
ax.set_yticks([-1, 0, 1])
ax.set_ylim(-1.5, 1.5)

ax.annotate('peak', xy=(25,1),
                xytext=(30, 1.5),
                arrowprops=dict(facecolor='black', headwidth=6, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')
ax.annotate('valley', xy=(75,-1),
                xytext=(80, -0.5),
                arrowprops=dict(facecolor='black', headwidth=6, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')
ax.plot(x,y, label='one')
ax.legend()
```
